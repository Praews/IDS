```python
a = np.arange(20).reshape(5,4)
df = pd.DataFrame(a,index='a b c d e'.split(),columns = 'V W X Y'.split())
df
```

```python
#จงสร้าง DataFrame(เฉลยอยู่ล่างสุด)
```

```python
df['Z'] = [1,2,3,4,5]
df
```

```python
#จงเพิ่ม Columns Z ลงไป กำหนดค่าเป็น 1,2,3,4,5
```

```python
df.loc['f'] = [6,7,8,9,10]
df
```

```python
#จงเพิ่ม Index f ลงไป กำหนดค่าเป็น 6,7,8,9,10
```

```python
len(df.columns)
```

```python
#จงแสดงจำนวน Columns ทั้งหมด
```

```python
df.loc['c'] =[8,0,0,0,3]
df.loc['d'] =[12,0,0,0,4]
df
```

```python
#เปลี่ยนตัวเลข 6 ตัวตรงกลางให้เป็น 0
```

```python
df['Sum'] = df['Y']+df['Z']
df
```

```python
#สร้าง Columns 'Sum' มีค่าเป็นผลรวมของ 'Y' และ 'Z'
```

```python
df.drop('Sum',axis=1,inplace = True)
df.drop('c',axis=0,inplace = True)
df
```

```python
#ทำการลบ Columns 'Sum' และ Index 'c' ออก และแสดงผลลัพธ์ออกมา (inplace = True)
```

```python
df.loc['e']
```

```python
#จงแสดงค่าของ Index 'e'
```

```python
df.rename(index = {'d':'c','e':'d','f':'e'},inplace = True)
df
```

```python
#จงเปลี่ยนชื่อ Index 'd' เป็น 'c' , 'e' เป็น 'd' และ 'f' เป็น 'e' และแสดงผลลัพธ์ออกมา (inplace = True) 
df.rename(index = {'d':'c','e':'d','f':'e'},inplace = True)
df
```

```python
df = pd.concat([df,df])
df
```

```python
#จงนำ DataFrame จากข้อก่อนหน้ามารวมกันให้ได้ผลลัพธ์แบบด้นล่าง
```

```python
df = pd.concat([df,df],axis = 1)
df
```
